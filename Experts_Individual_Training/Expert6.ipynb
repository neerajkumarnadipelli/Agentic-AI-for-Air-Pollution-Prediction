```python
# Expert 6: Quantum O3 Predictor - FIXED VARIABLE SCOPE
# Using TensorFlow SavedModel format for subclassed models

import os
os.environ['CUDA_VISIBLE_DEVICES'] = ''  # comment out to enable GPU

import time
import json
import numpy as np
import pennylane as qml
import tensorflow as tf
from tensorflow.keras import layers, Model, optimizers, callbacks
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# ----------------------------
# CONFIG
# ----------------------------
DATA_DIR = r"C:\Users\NNadi\Downloads\Air_pollution_Agentic_AI\datasets"
OUTPUT_DIR = r"C:\Users\NNadi\Downloads\Air_pollution_Agentic_AI\models\expert6_output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

SEQ_LEN = 5
PATCH_SIZE = 10
CHANNELS = 24
PRED_BAND_IDX = 5  # Changed from 4 to 5 for O3
NUM_QUBITS = 4
Q_DEPTH = 2
EPOCHS = 100
BATCH_SIZE = 4
LR = 1e-3

np.random.seed(42)
tf.random.set_seed(42)

print("TensorFlow:", tf.__version__)
print("PennyLane:", qml.__version__)

# ----------------------------
# Load datasets
# ----------------------------

def load_npy(name):
    path = os.path.join(DATA_DIR, name)
    if not os.path.exists(path):
        raise FileNotFoundError(path)
    return np.load(path)

X_train = load_npy("X_train.npy")
y_train = load_npy("y_train.npy")
X_val = load_npy("X_val.npy")
y_val = load_npy("y_val.npy")
X_test = load_npy("X_test.npy")
y_test = load_npy("y_test.npy")

print("Loaded shapes:")
print(" X_train:", X_train.shape)
print(" y_train:", y_train.shape)
print(" X_val  :", X_val.shape)
print(" y_val  :", y_val.shape)
print(" X_test :", X_test.shape)
print(" y_test :", y_test.shape)

# Prepare O3 targets
def extract_o3(y):
    o3 = y[..., PRED_BAND_IDX]  # Changed from CO to O3
    return o3[..., None].astype(np.float32)

y_train_o3 = extract_o3(y_train)  # Changed from y_train_co to y_train_o3
y_val_o3 = extract_o3(y_val)      # Changed from y_val_co to y_val_o3
y_test_o3 = extract_o3(y_test)    # Changed from y_test_co to y_test_o3

print("O3 target shapes:", y_train_o3.shape, y_val_o3.shape, y_test_o3.shape)

# ----------------------------
# Quantum-Enhanced Layer (Saves properly)
# ----------------------------

class QuantumEnhancedLayer(layers.Layer):
    """Custom layer that incorporates quantum-inspired operations"""
    
    def __init__(self, units, **kwargs):
        super().__init__(**kwargs)
        self.units = units
        
    def build(self, input_shape):
        # Quantum-inspired weights (complex numbers represented as 2D)
        self.quantum_weights = self.add_weight(
            shape=(input_shape[-1], self.units, 2),  # Real and imaginary parts
            initializer='random_normal',
            trainable=True,
            name='quantum_weights'
        )
        self.bias = self.add_weight(
            shape=(self.units,),
            initializer='zeros',
            trainable=True,
            name='bias'
        )
        super().build(input_shape)
    
    def call(self, inputs):
        # Quantum-inspired transformation
        real_part = tf.matmul(inputs, self.quantum_weights[:, :, 0])
        imag_part = tf.matmul(inputs, self.quantum_weights[:, :, 1])
        
        # Magnitude (quantum probability amplitude inspired)
        output = tf.sqrt(real_part**2 + imag_part**2) + self.bias
        return tf.nn.relu(output)  # Non-linearity
    
    def get_config(self):
        config = super().get_config()
        config.update({"units": self.units})
        return config

def create_expert6_quantum_enhanced():
    """Expert 6 with quantum-enhanced layers (saves properly)"""
    model = tf.keras.Sequential([
        layers.InputLayer(input_shape=(SEQ_LEN, PATCH_SIZE, PATCH_SIZE, CHANNELS)),
        
        # Spatio-temporal processing
        layers.BatchNormalization(),
        layers.Conv3D(16, (3, 3, 3), activation='relu', padding='same'),
        layers.Conv3D(32, (2, 2, 2), activation='relu', padding='same'),
        layers.GlobalAveragePooling3D(),
        
        # Quantum-enhanced processing
        QuantumEnhancedLayer(64),
        layers.Dropout(0.2),
        QuantumEnhancedLayer(32),
        layers.BatchNormalization(),
        
        # Spatial reconstruction
        layers.Dense(PATCH_SIZE * PATCH_SIZE * 16, activation='relu'),
        layers.Reshape((PATCH_SIZE, PATCH_SIZE, 16)),
        
        # Final refinement
        layers.Conv2D(32, 3, activation='relu', padding='same'),
        layers.Conv2D(16, 3, activation='relu', padding='same'),
        layers.Conv2D(8, 3, activation='relu', padding='same'),
        layers.Conv2D(1, 3, activation='linear', padding='same')
    ], name="Expert6_QuantumEnhanced_O3")  # Changed name from CO to O3
    
    return model

# ----------------------------
# Training Setup
# ----------------------------

def create_expert6_model():
    """Create Expert 6 model"""
    model = create_expert6_quantum_enhanced()
    
    optimizer = optimizers.Adam(learning_rate=LR)
    
    model.compile(
        optimizer=optimizer,
        loss='mse',
        metrics=['mae']
    )
    
    return model

def create_expert6_callbacks():
    """Create callbacks for training"""
    
    # Use TensorFlow SavedModel format
    checkpoint_cb = callbacks.ModelCheckpoint(
        os.path.join(OUTPUT_DIR, "expert6_best_model"),  # Changed from expert5 to expert6
        monitor='val_loss',
        save_best_only=True,
        save_weights_only=False,
        save_format='tf',  # Use TensorFlow format
        mode='min',
        verbose=1
    )
    
    lr_scheduler = callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-6,
        verbose=1
    )
    
    early_stopping = callbacks.EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    )
    
    return [checkpoint_cb, lr_scheduler, early_stopping]

# ----------------------------
# Training Execution
# ----------------------------

def train_expert6():
    """Complete training pipeline for Expert 6"""
    print("\n" + "="*60)
    print("TRAINING EXPERT 6: Quantum-Enhanced O3 Predictor")  # Changed from CO to O3
    print("="*60)
    
    start_time = time.time()
    
    # Create model
    print("Creating Expert 6 model...")  # Changed from Expert 5 to Expert 6
    expert6 = create_expert6_model()  # Changed from expert5 to expert6
    
    print("\nModel Architecture:")
    expert6.summary()
    
    # Create callbacks
    callbacks_list = create_expert6_callbacks()  # Changed from expert5 to expert6
    
    print(f"\nTraining details:")
    print(f"  Samples: {len(X_train)} training, {len(X_val)} validation")
    print(f"  Input shape: ({SEQ_LEN}, {PATCH_SIZE}, {PATCH_SIZE}, {CHANNELS})")
    print(f"  Target shape: ({PATCH_SIZE}, {PATCH_SIZE}, 1)")
    print(f"  Model type: Quantum-Enhanced Sequential")
    print(f"  Epochs: {EPOCHS}, Batch size: {BATCH_SIZE}")
    
    # Train model
    print("\nStarting training...")
    history = expert6.fit(  # Changed from expert5 to expert6
        X_train, y_train_o3,  # Changed from y_train_co to y_train_o3
        validation_data=(X_val, y_val_o3),  # Changed from y_val_co to y_val_o3
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        callbacks=callbacks_list,
        verbose=1,
        shuffle=True
    )
    
    training_time = time.time() - start_time
    print(f"\nTraining completed in {training_time:.2f} seconds")
    
    return expert6, history, training_time  # RETURN training_time

# ----------------------------
# Evaluation
# ----------------------------

def evaluate_expert6(model, X_test, y_test):  # Changed from expert5 to expert6
    """Comprehensive evaluation of Expert 6"""  # Changed from Expert 5 to Expert 6
    print("\n" + "="*60)
    print("EVALUATING EXPERT 6")  # Changed from Expert 5 to Expert 6
    print("="*60)
    
    # Predictions
    y_pred = model.predict(X_test, batch_size=BATCH_SIZE, verbose=1)
    
    # Flatten for metric calculation
    y_true_flat = y_test.reshape(-1)
    y_pred_flat = y_pred.reshape(-1)
    
    # Remove any invalid values
    mask = ~(np.isnan(y_true_flat) | np.isnan(y_pred_flat) | np.isinf(y_true_flat) | np.isinf(y_pred_flat))
    y_true_flat = y_true_flat[mask]
    y_pred_flat = y_pred_flat[mask]
    
    # Regression metrics
    mse = mean_squared_error(y_true_flat, y_pred_flat)
    mae = mean_absolute_error(y_true_flat, y_pred_flat)
    r2 = r2_score(y_true_flat, y_pred_flat)
    
    print("\nðŸ“Š PERFORMANCE METRICS:")
    print(f"  MSE:  {mse:.4f}")
    print(f"  MAE:  {mae:.4f}")
    print(f"  RÂ²:   {r2:.4f}")
    
    # Distribution analysis
    print("\nðŸ“ˆ DISTRIBUTION ANALYSIS:")
    print(f"  True - Min: {y_true_flat.min():.2f}, Max: {y_true_flat.max():.2f}, Mean: {y_true_flat.mean():.2f}")
    print(f"  Pred - Min: {y_pred_flat.min():.2f}, Max: {y_pred_flat.max():.2f}, Mean: {y_pred_flat.mean():.2f}")
    
    return {
        'mse': mse,
        'mae': mae, 
        'r2': r2,
        'predictions': y_pred,
        'y_true': y_true_flat,
        'y_pred': y_pred_flat
    }

# ----------------------------
# Visualization
# ----------------------------

def plot_expert6_results(history, evaluation_results):  # Changed from expert5 to expert6
    """Create visualization for Expert 6"""  # Changed from Expert 5 to Expert 6
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Loss plot
    axes[0,0].plot(history.history['loss'], label='Training Loss')
    axes[0,0].plot(history.history['val_loss'], label='Validation Loss')
    axes[0,0].set_title('Expert 6: Training and Validation Loss')  # Changed from Expert 5 to Expert 6
    axes[0,0].set_xlabel('Epoch')
    axes[0,0].set_ylabel('Loss')
    axes[0,0].legend()
    axes[0,0].grid(True)
    
    # MAE plot
    axes[0,1].plot(history.history['mae'], label='Training MAE')
    axes[0,1].plot(history.history['val_mae'], label='Validation MAE')
    axes[0,1].set_title('Expert 6: Training and Validation MAE')  # Changed from Expert 5 to Expert 6
    axes[0,1].set_xlabel('Epoch')
    axes[0,1].set_ylabel('MAE')
    axes[0,1].legend()
    axes[0,1].grid(True)
    
    # Prediction vs True scatter
    y_true = evaluation_results['y_true']
    y_pred = evaluation_results['y_pred']
    
    axes[1,0].scatter(y_true, y_pred, alpha=0.5, s=1)
    axes[1,0].plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', lw=2)
    axes[1,0].set_xlabel('True O3')  # Changed from CO to O3
    axes[1,0].set_ylabel('Predicted O3')  # Changed from CO to O3
    axes[1,0].set_title(f'Predictions vs True (RÂ² = {evaluation_results["r2"]:.3f})')
    axes[1,0].grid(True)
    
    # Error distribution
    errors = y_pred - y_true
    axes[1,1].hist(errors, bins=50, alpha=0.7, edgecolor='black')
    axes[1,1].axvline(x=0, color='r', linestyle='--')
    axes[1,1].set_xlabel('Prediction Error')
    axes[1,1].set_ylabel('Frequency')
    axes[1,1].set_title('Prediction Error Distribution')
    axes[1,1].grid(True)
    
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, 'expert6_results.png'), dpi=300, bbox_inches='tight')  # Changed from expert5 to expert6
    plt.show()

# ----------------------------
# Model Saving
# ----------------------------

def save_expert6_metadata(model, history, evaluation_results, training_time):  # Changed from expert5 to expert6
    """Save metadata and model for Expert 6"""  # Changed from Expert 5 to Expert 6
    
    metadata = {
        'expert_id': 6,  # Changed from 5 to 6
        'expert_name': 'QuantumEnhanced_O3_Predictor',  # Changed from CO to O3
        'architecture': {
            'type': 'Quantum-Enhanced Sequential',
            'input_shape': [SEQ_LEN, PATCH_SIZE, PATCH_SIZE, CHANNELS],
            'output_shape': [PATCH_SIZE, PATCH_SIZE, 1],
            'quantum_inspired': True
        },
        'training': {
            'epochs_trained': len(history.history['loss']),
            'final_train_loss': float(history.history['loss'][-1]),
            'final_val_loss': float(history.history['val_loss'][-1]),
            'training_time_seconds': float(training_time),
            'batch_size': BATCH_SIZE,
            'learning_rate': LR
        },
        'performance': {
            'test_mse': float(evaluation_results['mse']),
            'test_mae': float(evaluation_results['mae']),
            'test_r2': float(evaluation_results['r2'])
        },
        'target_band': 'O3',  # Changed from CO to O3
        'saved_format': 'TensorFlow SavedModel',
        'timestamp': time.strftime('%Y-%m-%d %H:%M:%S')
    }
    
    # Save metadata
    with open(os.path.join(OUTPUT_DIR, 'expert6_metadata.json'), 'w') as f:  # Changed from expert5 to expert6
        json.dump(metadata, f, indent=2)
    
    # Save final model in TensorFlow format
    model.save(os.path.join(OUTPUT_DIR, 'expert6_final_model'), save_format='tf')  # Changed from expert5 to expert6
    
    # Also save weights for compatibility
    model.save_weights(os.path.join(OUTPUT_DIR, 'expert6_weights.h5'))  # Changed from expert5 to expert6
    
    print(f"\nðŸ’¾ Expert 6 saved successfully!")  # Changed from Expert 5 to Expert 6
    print(f"   Model: {OUTPUT_DIR}/expert6_final_model/ (TensorFlow format)")  # Changed from expert5 to expert6
    print(f"   Weights: {OUTPUT_DIR}/expert6_weights.h5")  # Changed from expert5 to expert6
    print(f"   Metadata: {OUTPUT_DIR}/expert6_metadata.json")  # Changed from expert5 to expert6

# ----------------------------
# Model Loading Function
# ----------------------------

def load_expert6_model():  # Changed from expert5 to expert6
    """Load trained Expert 6 model"""  # Changed from Expert 5 to Expert 6
    model_path = os.path.join(OUTPUT_DIR, 'expert6_final_model')  # Changed from expert5 to expert6
    if os.path.exists(model_path):
        return tf.keras.models.load_model(model_path, custom_objects={'QuantumEnhancedLayer': QuantumEnhancedLayer})
    else:
        print("Model not found. Please train first.")
        return None

# ----------------------------
# Main Execution
# ----------------------------

if __name__ == "__main__":
    print("ðŸš€ STARTING EXPERT 6: Quantum-Enhanced O3 Predictor")  # Changed from CO to O3
    
    try:
        # Step 1: Train Expert 6
        expert6_model, training_history, training_time = train_expert6()  # Changed from expert5 to expert6
        
        # Step 2: Evaluate Expert 6
        evaluation_results = evaluate_expert6(expert6_model, X_test, y_test_o3)  # Changed from expert5 to expert6 and y_test_co to y_test_o3
        
        # Step 3: Visualize results
        plot_expert6_results(training_history, evaluation_results)  # Changed from expert5 to expert6
        
        # Step 4: Save model and metadata
        save_expert6_metadata(expert6_model, training_history, evaluation_results, training_time)  # Changed from expert5 to expert6
        
        print("\nâœ… EXPERT 6 COMPLETED SUCCESSFULLY!")  # Changed from Expert 5 to Expert 6
        print("ðŸ“ Model ready for ensemble stacking!")
        
        # Test loading the model
        print("\nðŸ§ª Testing model loading...")
        loaded_model = load_expert6_model()  # Changed from expert5 to expert6
        if loaded_model:
            print("âœ… Model loaded successfully for ensemble integration!")
            
    except Exception as e:
        print(f"\nâŒ ERROR in Expert 6: {e}")  # Changed from Expert 5 to Expert 6
        import traceback
        traceback.print_exc()
```
